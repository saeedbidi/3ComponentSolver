```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import copy

folder_path = '/Users/saeedbidi/AMReX/Dev/KAP-DEV-CT-Generalised/Exec/NewForThesis/3ComponentSolver/3KSP-DEV-CT/2DShockBubbleGallbladderCT/PythonCode/SelectedTimesPTWall/'
PTwall = []

file_list = os.listdir(folder_path)

for file in file_list:
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        df = df.drop(columns=['Points_1','Points_2'])
        df = df[['Points_0','P (GPa)','T3']]
        df = df.rename(columns={'Points_0':'X'})
        dfghar = copy.deepcopy(df)
        dfghar['X'] = abs (dfghar['X'])
        dfTot = pd.concat([dfghar,df])
        dfTot = dfTot.sort_values(by='X')
        PTwall.append(dfTot)

# Create separate line plots for each DataFrame
for index, df in enumerate(PTwall):
    # Extract x and y data for the line plot (replace 'x_column' and 'y_column' with your actual column names)
    x_data = df['X']
    y_data = df['T3']
    
    # Plot the data for each DataFrame, using a different label
    plt.plot(x_data, y_data, label=f'CSV {index + 1}')

# Customize the plot (add title, labels, legend, etc. as needed)
# plt.title('Liquid temperature along the wall at different times')
plt.xlabel('r-Axis',fontsize=14)
plt.ylabel('Liquid temperature along the wall (K)',fontsize=14)

# Increase the font size of tick labels on both axes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend()
    
    # Save or display the plot (optional)
    # plt.savefig(f'line_plot_csv_{index + 1}.png')  # Save each plot to a file
    # plt.show()  # Uncomment to display the plot interactively
output_filename = 'T.png'  # Change the filename and format as needed
plt.savefig(output_filename, dpi=600, bbox_inches='tight')  # 300 DPI is a decent quality

# Show all the separate line plots
plt.show()




```

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

####################################################################################
# Rigid Wall Case
####################################################################################

# Step 1: Read CSV files and merge them
path_to_csvs = '/Users/saeedbidi/AMReX/Dev/KAP-DEV-CT-Generalised/Exec/NewForThesis/3ComponentSolver/3KSP-DEV-CT/2DShockBubbleGallbladderCT/PythonCode/wall-data/'

all_dataframes = []

file_list = os.listdir(path_to_csvs)

for file in file_list:
    if file.endswith('.csv'):
        file_path = os.path.join(path_to_csvs, file)
        df = pd.read_csv(file_path)
        all_dataframes.append(df)

# Concatenate all dataframes into one
merged_data = pd.concat(all_dataframes, axis=0, ignore_index=True)
# display(merged_data)
# Group by 'Time' and 'Points:0', then average the temperature 'Tm'
grouped = merged_data.groupby(['Time', 'Points:0'])['Tm'].mean().reset_index()
grouped['Time'] = grouped['Time']*1647/0.05e-3-15
grouped['Points:0'] = grouped['Points:0']*1000
# grouped.rename(columns={'Tm':'T$_{l}$'})
 

# Keep the option to simply drop the duplicates
# Drop duplicates based on 'Time' and 'Points:0'
# merged_data.drop_duplicates(subset=['Time', 'Points:0'], inplace=True)

# Step 2: Pivot the grouped data to get heatmap structure
heatmap_data = grouped.pivot(index='Time', columns='Points:0', values='Tm')
heatmap_data.interpolate(method='quadratic', axis=1, inplace=True)  # Interpolate horizontally
heatmap_data.interpolate(method='quadratic', axis=0, inplace=True)  # Interpolate vertically
heatmap_data_transposed = heatmap_data.transpose()

# Plot the heatmap with swapped axes
plt.figure(figsize=(10, 8))
im = plt.imshow(heatmap_data_transposed, aspect='auto', cmap='inferno', origin='lower', extent=[heatmap_data.index.min(), heatmap_data.index.max(), heatmap_data.columns.min(), heatmap_data.columns.max()])
colorbar = plt.colorbar(im, label='T$_{l}$ (K)')
colorbar.ax.tick_params(labelsize=14)  # Set the font size for colorbar labels
colorbar.set_label(label='T$_{l}$ (K)', fontsize=18)  # Adjust the fontsize as needed

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('t$^{*}$',fontsize=14)
plt.ylabel('Wall axis (mm)',fontsize=14)
plt.xlim(0.0, max(grouped['Time']))
# plt.title('Temperature Heatmap')
plt.savefig("wall-temperature-map.png", bbox_inches='tight', dpi=720)
plt.show()

# # Step 3: Plot heatmap
# plt.figure(figsize=(10, 8))
# im = plt.imshow(heatmap_data, aspect='auto', cmap='inferno', origin='lower', extent=[heatmap_data.columns.min(), heatmap_data.columns.max(), heatmap_data.index.min(), heatmap_data.index.max()])
# # im = plt.imshow(heatmap_data, aspect='auto', cmap='inferno', norm=mpl.colors.LogNorm(vmin=290, vmax=360), origin='lower')
# plt.colorbar(im, label='Temperature (Tm)')
# plt.xlabel('x (m)')
# plt.ylabel(r'Time ($\mu$s)')
# plt.ylim(0.0, 1.1e-6)
# plt.title('Temperature Heatmap')
# plt.savefig("wall-temperature-map.png", bbox_inches='tight', dpi=720)
# plt.show()
```
